10  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution?



                      import numpy as np
                      import matplotlib.pyplot as plt

                      # Step 1: Define the parameters
                      population_size = 100000
                      sample_size = 30
                      num_samples = 1000

                      # Step 2: Generate data from a non-normal (exponential) distribution
                      population = np.random.exponential(scale=2.0, size=population_size)

                      # Step 3: Collect sample means
                      sample_means = []
                      for _ in range(num_samples):
                          sample = np.random.choice(population, size=sample_size, replace=False)
                          sample_means.append(np.mean(sample))

                      # Step 4: Plot the original (non-normal) distribution
                      plt.figure(figsize=(14, 6))

                      plt.subplot(1, 2, 1)
                      plt.hist(population, bins=50, density=True, color='tomato', edgecolor='black')
                      plt.title("Original Exponential Distribution")
                      plt.xlabel("Value")
                      plt.ylabel("Density")
                      plt.grid(True, linestyle='--', alpha=0.6)

                      # Step 5: Plot the distribution of sample means (CLT effect)
                      plt.subplot(1, 2, 2)
                      plt.hist(sample_means, bins=30, density=True, color='skyblue', edgecolor='black')
                      plt.title(f"Sampling Distribution of Mean (n={sample_size})\nCentral Limit Theorem Demonstration")
                      plt.xlabel("Sample Mean")
                      plt.ylabel("Density")
                      plt.grid(True, linestyle='--', alpha=0.6)

                      plt.tight_layout()
                      plt.show()



     Output:


              Left Plot: The original exponential distribution — skewed heavily to the right.

16  Create a confidence interval for a dataset using Python and interpret the result?



                    import numpy as np
                    from scipy import stats

                    # Sample data (e.g., test scores)
                    data = np.array([52, 49, 51, 53, 54, 48, 50, 47, 52, 51,
                                    49, 50, 53, 54, 48, 50, 51, 52, 50, 49])

                    # Sample size
                    n = len(data)

                    # Sample mean and standard deviation
                    mean = np.mean(data)
                    std_dev = np.std(data, ddof=1)  # ddof=1 for sample standard deviation

                    # Confidence level
                    confidence = 0.95
                    alpha = 1 - confidence

                    # Degrees of freedom
                    df = n - 1

                    # t-critical value
                    t_crit = stats.t.ppf(1 - alpha/2, df)

                    # Margin of error
                    margin_of_error = t_crit * (std_dev / np.sqrt(n))

                    # Confidence interval
                    lower_bound = mean - margin_of_error
                    upper_bound = mean + margin_of_error

                    print(f"Sample mean = {mean:.2f}")
                    print(f"Sample standard deviation = {std_dev:.2f}")
                    print(f"{int(confidence*100)}% confidence interval for the mean: ({lower_bound:.2f}, {upper_bound:.2f})")

              t_crit = stats.t.ppf(1 - (1-confidence)/2, df)

              lower_bound = sample_mean - margin_error
              upper_bound = sample_mean + margin_error

              # Step 7: Output results
              print(f"Sample Mean: {sample_mean:.2f}")

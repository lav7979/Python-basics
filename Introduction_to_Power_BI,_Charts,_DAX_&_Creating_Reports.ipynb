      Power BI Desktop

              Power BI is a business analytics tool developed by Microsoft that allows users to collect, analyze, visualize, and share data from various sources. It helps organizations make data-driven decisions through interactive reports and dashboards.



        Key Components of the Power BI Ecosystem
        1. Power BI Desktop

        It is a Windows application used by data analysts and developers to create reports and data models.

        Users can import data from multiple sources, transform it using Power Query, and design interactive visualizations.

        Reports created in Power BI Desktop can be published to the Power BI Service for sharing and collaboration.


        2. Power BI Service (Power BI Online)

        It is a cloud-based platform where users can view, share, and collaborate on Power BI reports and dashboards.

        It allows real-time data updates, dashboard sharing, and role-based access control.

        The service provides workspace environments for teams to manage and distribute analytics content.


            3. Power BI Mobile

            Power BI Mobile is a mobile application available for Android, iOS, and Windows devices.

            It enables users to access and interact with dashboards and reports on the go.

            It supports real-time notifications and touch-friendly visuals for easy navigation.

            4. Power BI Gateway

            Power BI Gateway acts as a bridge between on-premises data sources and the Power BI Service.

            It allows secure data transfer without moving data to the cloud.

            There are two types:

            Personal Gateway: For individual use.

            Enterprise Gateway: For large organizations to connect multiple data sources securely.

        1. Pie Chart vs Donut Chart
      Feature	Pie Chart	Donut Chart
      Shape	Circular chart divided into slices	Similar to Pie Chart but with a hole in the center
      Purpose	Shows percentage contribution of each category to a whole	Shows part-to-whole relationship but allows space to show total or additional info in the center
      Best Used When	There are few categories (3–6) for clear visibility	You want to show proportions and display total or key metrics in the center
      Limitations	Can look cluttered with many categories	Slightly less space for slices, but more modern and flexible design



            When to Prefer:

            Use a Pie Chart for simple, direct comparisons.

            Use a Donut Chart when you want a cleaner, modern look with a total value in the middle.

            Example using Global_Superstore2.csv:

            Pie Chart: Show Sales by Region (e.g., East, West, Central, South).

            Donut Chart: Show Sales by Category (e.g., Furniture, Office Supplies, Technology) with Total Sales in the center.

            2. Bar Chart vs Column Chart
            Feature	Bar Chart	Column Chart
            Orientation	Horizontal bars	Vertical bars
            Best For	Comparing many categories or long labels	Showing data over time or sequential comparisons
            Readability	Easier to read for long text labels	Easier to interpret for time trends
            Focus	Focuses on categories	Focuses on trend or time period

            When to Prefer:

            Use a Bar Chart when comparing multiple categories with long names.

            Use a Column Chart when analyzing changes or growth over time.

            Example using Global_Superstore2.csv:

            Bar Chart: Compare Total Profit by Sub-Category (like Chairs, Phones, Storage, etc.).

            Column Chart: Show Monthly Sales Trend across 2023 to analyze performance over time.

   3 Explain the significance of:
● Star schema vs Snowflake schema
● Primary key vs Foreign key in relationships (Power BI)
Why is cardinality important?





          Star Schema vs Snowflake Schema
        Feature	Star Schema	Snowflake Schema
        Structure	A central fact table connected directly to dimension tables	Dimension tables are normalized (split into multiple related tables) forming a snowflake-like structure
        Complexity	Simple and easy to understand	More complex due to multiple joins
        Performance	Faster – fewer joins and better query performance	Slower – requires more joins between tables
        Storage	Takes more space (duplicate data in dimensions)	Less storage (no data redundancy)
        Best For	Used in Power BI and data marts for quick analytics and visualization	Used in data warehouses where data consistency and normalization are important





        Power BI prefers Star Schema because it simplifies relationships, improves query speed, and reduces data model complexity.



          Using the Global_Superstore2.csv dataset –

          Fact Table: Sales (contains Sales, Profit, Quantity, Discount)

          Dimension Tables: Products, Customers, Region, Date

          These directly connect to the fact table → forming a Star Schema.

          2. Primary Key vs Foreign Key in Relationships (Power BI)
          Concept	Primary Key	Foreign Key
          Definition	A unique identifier for each record in a table	A field in one table that refers to the Primary Key of another table
          Purpose	Ensures that each record is unique	Creates relationships between tables
          Example in Power BI	CustomerID in Customer Table (unique)	CustomerID in Sales Table (repeats for multiple sales)
          Role in Power BI	Used in dimension tables	Used in fact tables to link data



          In Power BI:

          Relationships are built between primary and foreign keys, enabling filtering and cross-table analysis (e.g., filtering sales data by customer or region).





        Cardinality defines the type of relationship between two tables in Power BI.
        It determines how data flows and how filters are applied between tables.

        Type of Cardinality	Meaning	Example
        One-to-Many (1: )*	One record in one table relates to many in another (most common)	One Customer → Many Sales
        Many-to-One (*:1)	Reverse of above (Power BI auto-adjusts direction)	Many Sales → One Customer
        One-to-One (1:1)	Each record in one table matches exactly one in another	One Employee → One Login
        Many-to-Many (:)	Both tables have repeating values	Products sold in Multiple Stores and vice versa



          Ensures correct data relationships and accurate aggregations.

          Controls filter direction and data model performance.

          Helps Power BI avoid duplicate counting or data mismatches in visuals.







  4  Differentiate between:
● Calculated column vs Measure
Also, define Row context and Filter context with simple examples?





   
              Definition	A new column created in a table using DAX formulas; calculated row by row	A calculation performed on aggregated data; computed only when used in a visual
              Storage	Stored in the data model (takes memory space)	Not stored — calculated on the fly (more efficient)
              Evaluation	Evaluated for each individual row	Evaluated based on filter context in reports
              Use Case	When you need a value for each row (e.g., Profit per row)	When you need aggregated summaries (e.g., Total Profit, Average Sales)
              Example (using Global_Superstore2.csv)	Profit Margin = [Profit] / [Sales] → adds a new column for each sale	Total Sales = SUM(Sales[Sales]) → calculates total sales dynamically in visuals
              Performance	Slower for large datasets (takes model space)	Faster and more optimized for analysis


          2. Row Context vs Filter Context
          Concept	Row Context	Filter Context
          Definition	The context that exists for each individual row in a table	The context created by filters, slicers, or visuals in a report
          Used In	Calculated Columns or iterating functions like SUMX, FILTER	Measures and report visuals
          Evaluation Level	Row-by-row	Aggregate level (based on user filters)
          Example (Global_Superstore2.csv)	Calculated Column: Profit Margin = [Profit] / [Sales] → For each row, margin is calculated using row context.	Measure: Total Sales = SUM(Sales[Sales]) → When you select “Region = East” in a visual, Power BI applies filter context to show total sales only for East.
          Analogy	Think of it as working on each record separately	Think of it as filtering what data is visible before calculation





 | Concept               | Meaning                             | Example                                          |
| --------------------- | ----------------------------------- | ------------------------------------------------ |
| **Calculated Column** | Fixed value added to each row       | Profit Margin per Order                          |
| **Measure**           | Dynamic value calculated in visuals | Total Sales by Region                            |
| **Row Context**       | Calculation per row                 | Row-level profit margin                          |
| **Filter Context**    | Calculation based on filters        | Total sales when filtering by Region or Category |








  5  What is the difference between a report and a dashboard in Power BI?



   |                                                                                                                          |
| ----------------- | ---------------------------------------------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------- |
| **Definition**    | A **detailed, multi-page visualization** created in **Power BI Desktop or Service** showing data from a single dataset | A **single-page, summary view** (often called a *canvas*) created in **Power BI Service** using visuals pinned from one or more reports |
| **Pages**         | Can have **multiple pages**                                                                                            | Always **one page** only                                                                                                                |
| **Data Sources**  | Based on **one dataset only**                                                                                          | Can combine **multiple datasets and reports**                                                                                           |
| **Interactivity** | Highly interactive – supports **slicers, filters, drill-through, and cross-highlighting**                              | Limited interactivity – mainly **view and navigate to underlying reports**                                                              |
| **Creation Tool** | Created using **Power BI Desktop** and then published to the service                                                   | Created **only in Power BI Service**                                                                                                    |
| **Purpose**       | For **detailed analysis** and exploration of data                                                                      | For **high-level monitoring** and quick insights                                                                                        |
| **Example**       | A **Sales Report** showing sales, profit, and customer data with filters for region and time                           | A **Sales Dashboard** summarizing key KPIs like *Total Sales*, *Profit Margin*, and *Top 5 Products* on one screen

          Sales % of Total =
          DIVIDE(
              SUM('Sales'[Sales_Amount]),
              CALCULATE(SUM('Sales'[Sales_Amount]), ALL('Sales'))

          DIVIDE(...)	Divides category sales by total sales safely (avoids divide-by-zero errors).
